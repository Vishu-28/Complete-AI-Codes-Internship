```python
# importing correct models

# importing linearSVC, KNeighborsClassifier, RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
```

### 3.2. Model Training and 3.3. Model Evaluation

```python
# create a dictornary of models
models = {
    'RFC': RandomForestClassifier(),
    'LSVC': LinearSVC(),
    'KNC': KNeighborsClassifier()
}
```

```python
# creating a function to fit (train) and score (evaluate) the models

def fit_and_score(model, x_train, x_test, y_train, y_test):
    """
    Fits (trains) and Scores (evaluates) the entered data on the models.
    Args:
    models: a dictornary of different Scikit-Learn ML models
    x_train: training data
    x_test: testing data
    y_train: labels associated with training data
    y_test: labels associated with testing data
    """
    # model training
    np.random.seed(42)
    trained_models = {}
    for name, model in models.items():
        model.fit(x_train, y_train)   
        trained_models[name] = model
    
    # model evaluation
    # setting the random seed
    np.random.seed(42)
    # creating an empty dictornary to store model scores
    model_scores = {}
    # looping through the models
    for name, model in models.items():
        # fitting the model to the training data
        model.fit(x_train, y_train)
        # scoring the model and storing the score in the dictornary
        model_scores[name] = model.score(x_test, y_test) 
    return model_scores
```

### 3.4. Model Comparision

```python
# testing the function
model_scores = fit_and_score(models, x_train, x_test, y_train, y_test)
model_scores
```

### 3.7. Cross-Validation

### Cross-Validation for KNeighborsClassifier

Cross-Validation is used to make sure that our data is not biased while training. 

Learning the parameters of a prediction function and testing it on the same data is a methodological mistake: a model that would just repeat the labels of the samples that it has just seen would have a perfect score but would fail to predict anything useful on yet-unseen data. This situation is called overfitting. To avoid it, it is common practice when performing a (supervised) machine learning experiment to hold out part of the available data as a test set X_test, y_test. 

[Scikit-Learn](https://scikit-learn.org/stable/modules/cross_validation.html)

```python
# cross-validation
from sklearn.model_selection import cross_val_score
np.random.seed(42)
knn_model = KNeighborsClassifier()
scores = cross_val_score(knn_model, x, y, cv=5)
scores
```

```python
scores.mean()*100
```

### 3.5. Model Hyperparameter-Tuning

there are 3 ways to perform Hyperparameter Tuning with Scikit-Learn
1. Manual
2. RandomSearchCV
3. GridSearchCV

#### 1. Manual Hyperparameter Tuning

```python
knn_model.get_params()
```

```python
# creating a list of train scores
train_scores = []
#creating a list of test scores
test_scores = []
# creating a list of different values for n_neighbors
neighbors = range(1, 21)
# setup KNN instance
knn_model = KNeighborsClassifier()
# looping through different values of n_neighbors 
for i in neighbors:
    knn_model.set_params(n_neighbors=i)
    # fitting the model
    knn_model.fit(x_train, y_train)
    # appending the training score to the train_scores list
    train_scores.append(knn_model.score(x_train, y_train))
    # appending the test score to the test_scores list
    test_scores.append(knn_model.score(x_test, y_test))
```

```python
train_scores
```

```python
test_scores
```

```python
plt.plot(neighbors, train_scores, label='Train Score')
plt.plot(neighbors, test_scores, label='Test Score')
plt.xticks(np.arange(1, 21, 1))
plt.legend()
plt.xlabel('No. of Neighbors')
plt.ylabel('Model Score')
plt.show()
```

#### 2. RandomSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
knn_rand_grid = {
    'n_neighbors': np.arange(1, 21),
    'weights': ['uniform', 'distance'],
    'p': [1, 2, 1.5],
}
```

```python
np.random.seed(42)
rs_knn = RandomizedSearchCV(
    KNeighborsClassifier(),
    param_distributions=knn_rand_grid,
    cv=5,
    n_iter=20,
    verbose=True
)

# fit the random hyperparameter search model
rs_knn.fit(x_train, y_train)
```

```python
rs_knn.best_params_
```

```python
rs_knn.score(x_test, y_test)*100
```

#### 3. GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_grid_grid = {
    'n_neighbors': np.arange(1, 21),
    'weights': ['uniform', 'distance'],
    'p': [1, 2, 1.5],
}
```

```python
np.random.seed(42)
gs_knn = GridSearchCV(
    KNeighborsClassifier(),
    param_grid=knn_grid_grid,
    cv=5,
    verbose=True
)

# fit the random hyperparameter search model
gs_knn.fit(x_train, y_train)
```

```python
gs_knn.best_params_
```

```python
gs_knn.score(x_test, y_test)*100
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
sns.set(font_scale=1)
def plot_conf_mat(y_test, y_preds):
    """
    Plots a confusion matrix using Seaborn's heatmap.
    Args:
    y_test: true labels
    y_preds: predicted labels
    """
    fig, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(confusion_matrix(y_true=y_test, y_pred=y_preds),
                     annot=True,
                     cmap='Blues',
                     fmt='d',
                     cbar=False)
    plt.xlabel('True Label')
    plt.ylabel('Predicted Label')
    plt.title('Confusion Matrix')
```

```python
plot_conf_mat(y_test=y_test, y_preds=y_preds)
```
